```python
rows = []

# Get rows for each species, cluster, and presence/absence boolean value

```

```python
species = df2['Latin name'].unique()
species
```

```python
labels = df2['Label'].unique()
labels
```

```python
for organism in species:
    for label in labels:
        rows.append({'Species': organism, 'Label': label, 'Present': False}) # Populate Present with False for now
```

```python
# rows
```

```python
# Check expected number of rows
expectedRows = len(species)*len(labels)
# Check that actual number of rows is equal to expected
actualRows = len(rows)

if expectedRows == actualRows:
    print(f"Rejoice! Actual equals expected. \nActual: {actualRows} \tExpected: {expectedRows}")
else:
    print(f"Go ahead and retry that. The actual does not equal expected.\nActual: {actualRows} \tExpected: {expectedRows}")
```

```python
# It is, so populate dataframe with False values for now
dfLabelBool = pd.DataFrame(rows)
dfLabelBool
```

```python
df2.rename(columns={'Species': 'Species_code', 'Latin name': 'Species','Cluster #': 'Cluster'}, inplace=True)
df2
```

```python
dfLabelBool['Present'] = dfLabelBool[['Species', 'Label']].merge(
    df2[['Species', 'Label']].drop_duplicates(),
    on=['Species', 'Label'],
    how='left',
    indicator=True
)['_merge'].eq('both')
```

```python
# Verify it worked by using ZNF777 as an example. There should be 72 species with ZNF777
dfLabelBool.loc[(dfLabelBool['Label']=='ZNF777')&(dfLabelBool['Present']==True)]
```

```python
# Drop NaN values
dfLabelBool = dfLabelBool.dropna()
```

```python
dfLabelBool
```

```python
df2.rename(columns={'Estimated evolutionary distance from human (MY)': 'timeFromHuman_MY', 'Common name': 'CommonName'}, inplace=True)

df2
```

```python
# Convert long to wide
df_znf_wide = dfLabelBool.pivot(index='Species', columns='Label', values='Present').reset_index()
df_znf_wide.columns.name = None
```

```python
df_znf_wide
```

```python
df_znf_wide = df_znf_wide.drop('noname', axis=1)
```

```python
df_znf_wide
```

```python
df_znf_wide = df_znf_wide.merge(
    df2[['Species', 'Order', 'Class', 'CommonName', 'timeFromHuman_MY']],
    on='Species',
    how='left'
)
df_znf_wide
```

```python
df_znf_wide = df_znf_wide.drop_duplicates()
df_znf_wide
```

```python
col0 = df_znf_wide.pop('Order')        # remove the column
df_znf_wide.insert(1, 'Order', col0)   # insert it at position 1 (0-based index)

col1 = df_znf_wide.pop('Class')
df_znf_wide.insert(2, 'Class', col1)

col2 = df_znf_wide.pop('CommonName')
df_znf_wide.insert(3, 'CommonName', col2)

col3 = df_znf_wide.pop('timeFromHuman_MY')
df_znf_wide.insert(4, 'timeFromHuman_MY', col3)

# Check df_wide
df_znf_wide
```

```python
df_znf_wide.to_csv("df_znf_wide.csv", index=False)
```
