```r
cluster_presence_monotremata_sorted_191 <- read.csv(
    "df_ornithorhynchus_anatinus_sorted.csv",
  header = TRUE,
  stringsAsFactors = FALSE
)
```

```r
cluster_presence_monotremata_sorted_191
```

```r
# keep order as in dataframe:
cluster_presence_monotremata_sorted_191$cluster <- factor(
    cluster_presence_monotremata_sorted_191$cluster,
    levels = rev(unique(cluster_presence_monotremata_sorted_191$cluster)),
    ordered=TRUE)
cluster_presence_monotremata_sorted_191$Species <- factor(
    cluster_presence_monotremata_sorted_191$Species,
    levels = unique(cluster_presence_monotremata_sorted_191$Species),
    ordered=TRUE)
```

```r
# set colors for plotting
colors <- c("lightgrey", "violetred4")

# plot
monotremata_clusters_191 <- 
    ggplot(cluster_presence_monotremata_sorted_191, aes(x=Species, y=factor(cluster), alpha=0.5)) +
    geom_tile(aes(fill=present)) + 
    scale_fill_manual(values=colors) +
    theme(axis.text.x = element_text(angle = 90, hjust = 1))
plot(monotremata_clusters_191)
```

```r
ggsave(monotremata_clusters_191, filename="monotremata_clusters_191_AllSortedClusters002.png", width=30, height=20, units = "cm")
```

```r
cluster_presence_mus_musculus_sorted_191 <- read.csv(
    "df_mus_musculus_sorted.csv",
  header = TRUE,
  stringsAsFactors = FALSE
)

# print(cluster_presence_mus_musculus_sorted_191)

# keep order as in dataframe:
cluster_presence_mus_musculus_sorted_191$cluster <- factor(
    cluster_presence_mus_musculus_sorted_191$cluster,
    levels = rev(unique(cluster_presence_mus_musculus_sorted_191$cluster)),
    ordered=TRUE)
cluster_presence_mus_musculus_sorted_191$Species <- factor(
    cluster_presence_mus_musculus_sorted_191$Species,
    levels = unique(cluster_presence_mus_musculus_sorted_191$Species),
    ordered=TRUE)

# set colors for plotting
colors <- c("lightgrey", "violetred4")

# plot
mus_musculus_clusters_191 <- 
    ggplot(cluster_presence_mus_musculus_sorted_191, aes(x=Species, y=factor(cluster), alpha=0.5)) +
    geom_tile(aes(fill=present)) + 
    scale_fill_manual(values=colors) +
    theme(axis.text.x = element_text(angle = 90, hjust = 1))
plot(mus_musculus_clusters_191)
```
